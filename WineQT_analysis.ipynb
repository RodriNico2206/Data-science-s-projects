# Introduction to the Datas

The dataset was generated using [Wine Quality dataset](https://www.kaggle.com/yasserh/wine-quality-dataset
). The task is to predict right Quality which is an ordinal column.

Columns description taken from this [link](https://archive.ics.uci.edu/dataset/186/wine+quality):

* Number of Attributes: 11 + output attribute, input variables (based on physicochemical tests):

* Fixed acidity `(tartaric acid -  g/dm3)`: most acids involved with wine or fixed or nonvolatile (do not evaporate readily).

* Volatile acidity `(acetic acid -  g/dm3)`: the amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste.

* Citric acid `(g/dm3)`: found in small quantities, citric acid can add "freshness" and flavor to wines.

* Residual sugar `(g/dm3)`: the amount of sugar remaining after fermentation stops, it's rare to find wines with less than 1 gram/liter and wines with greater than 45 grams/liter are considered sweet.

* Chlorides `(sodium chloride -  g/dm3)`: the amount of salt in the wine.

* Free sulfur dioxide `(mg/dm3)`: the free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion; it prevents microbial growth and the oxidation of wine.

* Total sulfur dioxide `(mg/dm3)`: amount of free and bound forms of S02; in low concentrations, SO2 is mostly undetectable in wine, but at free SO2 concentrations over 50 ppm, SO2 becomes evident in the nose and taste of wine.

* Density `(g/cm3)`: density of water is close to that of water depending on the percent alcohol and sugar content.

* pH: describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic); most wines are between 3-4 on the pH scale.

* Sulphates `(potassium sulphate -  g/dm3)`: a wine additive which can contribute to sulfur dioxide gas (S02) levels, wich acts as an antimicrobial and antioxidant.

* Alcohol `(% by volume)`: the percent alcohol content of the wine.

Output variable (based on sensory data):

* Quality (score between 0 and 10).

# Box and Whisker

Visualization and calculus of maximum, minimum, median, first and third quartile, as also outlier values of each variable from wine data set. It related with one box and whiskers plot of each a them.

```python
# Quartiles of alcohol's values grouped by 'Quality'
df.quartiles = pd.DataFrame(
    data = df.groupby('quality').quantile([0.25, 0.5, 0.75])
)

# Get a list of variables to plot
variables = [column for column in df.columns if column != 'quality']

# Calculate the number of rows and columns for the subplot grid
num_plots = len(variables)
num_rows = (num_plots - 1) // 3 + 1
num_cols = min(num_plots, 3)

# Create subplots for the charts
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 5*num_rows))

# Iterate over each variable and create a box and whisker plot in the corresponding subplot
for i, variable in enumerate(variables):
    ax = axes[i // num_cols, i % num_cols]
    sns.boxplot(y='quality', x=variable, data=df, ax=ax, orient = 'h')
    ax.set_title(f'Box and Whisker Plot for {variable.title()}')

# Hide empty subplots
if num_plots < num_rows * num_cols:
    for j in range(num_plots, num_rows * num_cols):
        axes[j // num_cols, j % num_cols].axis('off')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()

df.quartiles
```

#Outliers

For count of outliers from each this data set's column, it utilized the Interquartile Range (IQR) method that is typically used to detect outliers.

Although, isn't exact it provide us a insight about which variables possess a major outliers number





```python
# Function to count outliers using the IQR method
def count_outliers(column):
    if column.name != 'quality':  # Exclude the 'quality' column
        Q1 = column.quantile(0.25)
        Q3 = column.quantile(0.75)
        IQR = Q3 - Q1
        return ((column < (Q1 - 1.5 * IQR)) | (column > (Q3 + 1.5 * IQR))).sum()
    return 0

# Count outliers for each column
outliers_count = df.apply(count_outliers)

# Remove the 'quality' column from the outliers_count Series
outliers_count = outliers_count.drop('quality')

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(outliers_count.index, outliers_count.values)
plt.xlabel('Columns')
plt.ylabel('Number of Outliers')
plt.title('Count of Outliers in Each Column (excluding "quality")')
plt.xticks(rotation=45)

# Add value tags on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, height, ha='center', va='bottom')

plt.show()
```

#Frequency Distribution

Frequency distribution and frequency polygon for each variable of data set, where it 's possible observe mean, mode and bias what possess the distribution of each variable.

```python


# Get the number of columns in the DataFrame
num_columns = len(df.columns)

# Set the number of rows and columns for subplots (3 charts per row)
num_rows = (num_columns + 2) // 3  # Round up the number of rows to the nearest multiple of 3
num_cols = min(3, num_columns)   # Maximum 3 columns per row

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 6, num_rows * 4))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Plot frequency histograms for each variable and add distribution lines
for i, column in enumerate(df.columns):
    sns.histplot(df[column],
             ax=axes[i],
             kde=True,
             bins= round(math.sqrt( len(df) ) ),
             color='skyblue',
             edgecolor='black'
             )
    axes[i].set_title(f'Histogram of {column}')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

# Remove any empty subplots
for i in range(num_columns, num_rows * num_cols):
    fig.delaxes(axes[i])

# Adjust layout and spacing between subplots
plt.tight_layout()

# Display the histograms
plt.show()
```
