# Outliers treatment
It's using method Capping and Flooring and other techniques

```python
ul = np.percentile(df.n_hot_rooms, [99])[0]
```

```python
df.n_hot_rooms[df.n_hot_rooms > 3*ul ] = 3*ul
```

```python
df[(df.n_hot_rooms > ul) ]
```

```python
ll = np.percentile(df.rainfall, [1])[0]
```

```python
df.rainfall[df.rainfall < .3*ll ] = .3*ll
```

```python
df[(df.rainfall < ll) ]
```

# Columns

- Coef: shows coefficients's value for each one of independent variables. Therefore, it's indicating at what amount the response variable will increases whether i increase by one whichever independent variable
- std err: standard error for each of independent variables.
- t: associated to each independent variables.
- P>|t|: p value, indicates which are most significantly variables. it's saying us, which the have highest impact in our y variable


**NOTE:** <span style="color:red;">we must identify p value less that 5%, then we must analyze their coefficient and your sign for make bussines sense to all our dats</span>

```python
#define regression model
lm3 = LinearRegression()

#adjust model
lm3.fit(X_multi,y_multi)
```

# Note about attributes

<ul style="list-style-type: square;">
  <li>
    Intercept attribute: <span style="margin-left: 10px;">represent the value of the dependent variable when all independent variables are zero. In other words, it's the value of the target variable when all predictors are absent.</span>
  </li>
  <li>
    Coef attribute: <span style="margin-left: 10px;">represent $\beta_1$, $\beta_2$, etc, for each one our variables. In other words, it's representing the weights applied to each feature (independent variable) when predicting the target variable. Each coefficient corresponds to a specific feature</span>
  </li>
</ul>

# Test Train split



```python
from sklearn.model_selection import train_test_split
```

```python
#we create our variables of training and testing
X_train, X_test, y_train, y_test = train_test_split(X_multi, y_multi, test_size = 0.2, random_state = 0)
```

```python
#we check out rows and columns of our variables
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#define regression model
lm_a = LinearRegression()

#adjust model
lm_a.fit(X_train,y_train)
```

```python
#we create test dependent variable predicted y using this model
y_test_a = lm_a.predict(X_test)
```

```python
#Similarly, we create train dependent variable predicted y using this model
y_train_a = lm_a.predict(X_train)
```

```python
#we're getting r squared
from sklearn.metrics import r2_score
```

```python
#command help for to call r2_score function
r2_score?
```

```python
#y_test: value true; y_test_a: value predicted 
#we're getting value r power to 2 for test dependent variables 

r2_score(y_test, y_test_a,)
```

```python
#we're getting value r power to 2 for train dependent variables 

r2_score(y_train, y_train_a,)
```


**NOTE:** <span style="color:red;">we always must look at r power to 2 of our test data set, instead of train data set to assess the performance of model</span>
